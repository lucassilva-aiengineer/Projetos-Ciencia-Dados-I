```python
from typing import Dict, List
```

```python

def catalogo_codigos()-> Dict[int, str]:

  codigos = [601, 600, 1400, 2703, 700, 500, 4022, 502, 609, 1199]

  lista_codigos: Dict[int, str] = {}
  for codigo in codigos:

      df_string = df[df["Highest Offense Code"] == codigo].reset_index()
      string = df_string.iloc[0]["Highest Offense Description"]
      lista_codigos[codigo] = string


  return lista_codigos


# string.head(2)

# string.iloc[0]["Highest Offense Description"]
```

```python
crimes_registrados_dias_diferentes = catalogo_codigos()
```

```python
for codigo, crime in crimes_registrados_dias_diferentes.items():
    print(f"{codigo}: {crime}")
```
