```python
df["Occurred Date Time"] = pd.to_datetime(df["Occurred Date Time"], format= "%m/%d/%Y %H:%M")
```

```python
df["dia ocorrido"] = df["Occurred Date Time"].dt.day
df["mes ocorrido"] = df["Occurred Date Time"].dt.month
df["ano ocorrido"] = df["Occurred Date Time"].dt.year
df["hora ocorrido"] = df["Occurred Date Time"].dt.hour
df["data ocorrido"] = df["Occurred Date Time"].dt.date
```

```python
quantidades_diarias = df["dia ocorrido"].value_counts()
quantidades_diarias = quantidades_diarias.reset_index()
```

```python
quantidades_diarias.head(2)
```

```python
quantidades_diarias.columns
```

```python
plt.scatter(quantidades_diarias["dia ocorrido"],
            quantidades_diarias["count"],
            color= "green")

plt.title("Relação Crimes dia do mês")

plt.show()
```

```python
# Quais os dias da semana mais perigosos em austin.
# Eu acredito que seja o final de semana


df["dia semana"] = df["Occurred Date Time"].dt.day_name()
```

```python
df["dia semana"].head()
```

```python
contagem_dias_semana = df["dia semana"].value_counts()
contagem_dias_semana = contagem_dias_semana.reset_index()
```

```python
contagem_dias_semana.head()
```

```python
plt.bar(np.array([dia[:6] for dia in  list(contagem_dias_semana["dia semana"])]),
        contagem_dias_semana["count"],
        color= "lightgreen",
        edgecolor= 'black')


plt.title("Dia da semana")
plt.grid(True,
         alpha= 0.8)


plt.show()
```

```python
# Que tal vermos em porcentagem

plt.pie(contagem_dias_semana["count"],
        labels= contagem_dias_semana["dia semana"],
        autopct= "%.2f%%")

plt.title("Distribuição dos crimes cometidos em Austin")

plt.show()


# É clara a distribuição com poucas diferenças.

# Dia com maior número de casos: Friday, Sexta-Feira 15.22%.
# Dia com menor número de casos: Tuesday, Terça-Feira 13.77%.

# Será que isso se repete em outras cidades?
# Será que isso também acontece quanto ao registro das ocorrências?
```

```python
# Quantos porcento dos crimes são registrados em datas diferentes do ocorrido.

registro_mesmo_dia = df[df["data ocorrido"] == df["registro data"]]
registro_outro_dia = df[df["data ocorrido"] != df["registro data"]]
```

```python
# O mesmo dia

from matplotlib import pyplot as plt


plt.pie([len(registro_mesmo_dia), len(registro_outro_dia)],
        labels= ["Registro mesmo dia", "Registro outro dia"],
        autopct= "%.2f%%")

plt.show()
```

```python
# Até quanto tempo de diferença eu devo considerar importante?
# Quais os crimes que não são relatados no mesmo praso do ocorrido?
```

```python
crimes_registro_posterior =  registro_outro_dia["Highest Offense Code"].value_counts()
crimes_registro_posterior = crimes_registro_posterior.reset_index()

crimes_registro_mesmo_dia = registro_mesmo_dia["Highest Offense Code"].value_counts()
crimes_registro_mesmo_dia = crimes_registro_mesmo_dia.reset_index()
```

```python
crimes_registro_mesmo_dia["count"]
```

```python
plt.bar([str(codigo) for codigo in list(crimes_registro_mesmo_dia["Highest Offense Code"])[:10]],
        list(crimes_registro_mesmo_dia["count"])[:10],
        color= "lightgreen",
        edgecolor= "black")

plt.grid(linestyle= "--")

plt.title("Crimes que são registrados no dia do ocorrido")
plt.show()
```

```python
# O que os código significam.

df_codigos_a = set(list(crimes_registro_mesmo_dia["Highest Offense Code"].isin(list(crimes_registro_mesmo_dia["Highest Offense Code"])[0:10]))[:10])
```

```python
df_codigos_a
```

```python
# Os crimes que são registrados em datas diferentes do ocorrido.


plt.bar([str(codigo) for codigo in list(crimes_registro_posterior['Highest Offense Code'])[:10]],
        [count for count in list(crimes_registro_posterior["count"])[:10]],
        color= "lightgreen",
        edgecolor= "black")


plt.title("Crimes registrados em dias diferentes do ocorrido")
plt.grid(linestyle= "--")

plt.show()
```
